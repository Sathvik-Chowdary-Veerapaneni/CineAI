```python
import pandas as pd

def extract_movie_details(movie_entry):
    #One movie information
    movie_id = list(movie_entry.keys())[0]
    movie_data = movie_entry[movie_id]

    details = movie_data.get("details", {})

    # Extracting fields from details endpoint
    row = {
        "movie_id": details.get("id"),
        "title": details.get("title"),
        "overview": details.get("overview"),
        "release_date": details.get("release_date"),
        "runtime": details.get("runtime"),
        "original_language": details.get("original_language"),
        "popularity": details.get("popularity")
    }

    return pd.DataFrame([row])
```

### Extracting all the details endpoint movie information into details_df

```python
def get_all_details(root_dir):

    movies_list = []
    for year in range(2000, 2024):
        year_path = os.path.join(root_dir, str(year))
        if os.path.exists(year_path) and os.path.isdir(year_path):
            # Looping each JSON file from the year folder
            for filename in sorted(os.listdir(year_path)):
                if filename.endswith('.json'):
                    file_path = os.path.join(year_path, filename)
                    with open(file_path, 'r') as f:
                        data = json.load(f)

                    # Extracting movie details for each entry in the data
                    for movie_entry in data:
                        movie_id = list(movie_entry.keys())[0]
                        movie_data = movie_entry[movie_id]
                        details = movie_data.get("details", {})

                        row = {
                            "movie_id": details.get("id"),
                            "title": details.get("title"),
                            "overview": details.get("overview"),
                            "release_date": details.get("release_date"),
                            "runtime": details.get("runtime"),
                            "original_language": details.get("original_language"),
                            "popularity": details.get("popularity")
                        }

                        movies_list.append(row)

    details_df = pd.DataFrame(movies_list)
    return details_df
```

```python
def get_all_credits(root_dir):
    cast_list = []
    crew_list = []
    for year in range(2000, 2024):
        year_path = os.path.join(root_dir, str(year))
        if os.path.exists(year_path) and os.path.isdir(year_path):
            # Loop through each JSON file in that year folder
            for filename in sorted(os.listdir(year_path)):
                if filename.endswith('.json'):
                    file_path = os.path.join(year_path, filename)
                    with open(file_path, 'r') as f:
                        data = json.load(f)

                    # Extracting credits endpoint for each movie entry
                    for movie_entry in data:
                        movie_id = list(movie_entry.keys())[0]
                        movie_data = movie_entry[movie_id]
                        credits = movie_data.get("credits", {})
                        
                        # Extracting cast
                        for cast_member in credits.get("cast", []):
                            cast_list.append({
                                "movie_id": movie_id,
                                "cast_id": cast_member.get("id"),
                                "name": cast_member.get("name"),
                                "character": cast_member.get("character"),
                                "credit_id": cast_member.get("credit_id")
                            })
                        
                        # Extracting crew
                        for crew_member in credits.get("crew", []):
                            crew_list.append({
                                "movie_id": movie_id,
                                "crew_id": crew_member.get("id"),
                                "name": crew_member.get("name"),
                                "department": crew_member.get("department"),
                                "job": crew_member.get("job"),
                                "credit_id": crew_member.get("credit_id")
                            })

    cast_df = pd.DataFrame(cast_list)
    crew_df = pd.DataFrame(crew_list)
    return cast_df, crew_df
```

```python
def get_all_recommendations(root_dir):
    recommendations_list = []
    for year in range(2000, 2024):
        year_path = os.path.join(root_dir, str(year))
        if os.path.exists(year_path) and os.path.isdir(year_path):
            # Loop through each JSON file in that year folder
            for filename in sorted(os.listdir(year_path)):
                if filename.endswith('.json'):
                    file_path = os.path.join(year_path, filename)
                    with open(file_path, 'r') as f:
                        data = json.load(f)

                    # Extracting recommendations endpoint for each movie entry
                    for movie_entry in data:
                        movie_id = list(movie_entry.keys())[0]
                        movie_data = movie_entry[movie_id]
                        recommendations = movie_data.get("recommendations", {})
                        
                        # Extracting recommended movies
                        for rec in recommendations.get("results", []):
                            recommendations_list.append({
                                "movie_id": movie_id,
                                "recommended_movie_id": rec.get("id"),
                                "recommended_title": rec.get("title"),
                                "recommended_popularity": rec.get("popularity"),
                                "recommended_genre_ids": rec.get("genre_ids")
                            })

    recommendations_df = pd.DataFrame(recommendations_list)
    return recommendations_df
```

```python
def get_all_watch_providers(root_dir):
    watch_providers_list = []
    for year in range(2000, 2024):
        year_path = os.path.join(root_dir, str(year))
        # Check if the year directory exists
        if os.path.exists(year_path) and os.path.isdir(year_path):
            # Iterating over files in the year directory
            for filename in sorted(os.listdir(year_path)):
                if filename.endswith('.json'):
                    file_path = os.path.join(year_path, filename)
                    # Load JSON data
                    with open(file_path, 'r') as f:
                        data = json.load(f)

                    # Extracting watch_providers for each movie entry
                    for movie_entry in data:
                        # Each movie_entry is { "movie_id": { "details": {...}, "credits": {...}, ... } }
                        movie_id = list(movie_entry.keys())[0]
                        movie_data = movie_entry[movie_id]
                        watch_providers = movie_data.get("watch_providers", {})

                        # Access 'results' key to reach US-specific data
                        results = watch_providers.get("results", {})
                        # Verify 'results' is a dictionary before accessing 'US'
                        if not isinstance(results, dict):
                            # If results is not dict, skip this entry
                            continue

                        # Get US-specific watch provider data
                        us_data = results.get("US", {})
                        # Verify us_data is a dictionary
                        if not isinstance(us_data, dict):
                            # If US data is not dict, skip
                            continue

                        # Extract the link field
                        link = us_data.get("link")

                        # rent, buy, flatrate keys might not always be present, handle gracefully
                        for provider_type in ["rent", "buy", "flatrate"]:
                            providers = us_data.get(provider_type, [])
                            # Verify providers is a list
                            if not isinstance(providers, list):
                                continue

                            # Iterate over each provider entry
                            for p in providers:
                                watch_providers_list.append({
                                    "movie_id": movie_id,
                                    "provider_type": provider_type,
                                    "provider_name": p.get("provider_name"),
                                    "link": link
                                })

    # Convert collected data into a DataFrame
    watch_providers_df = pd.DataFrame(watch_providers_list)
    return watch_providers_df
```

```python
cast_df, crew_df = get_all_credits(root_dir)
recommendations_df = get_all_recommendations(root_dir)
watch_providers_df = get_all_watch_providers(root_dir)
```
